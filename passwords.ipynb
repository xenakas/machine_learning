### Creating features

#### Feature 1

- or three other features:  # of letters, # of digits, # of symbols

```python
def feature_1(inputString):
    valid_symbols = "!?.,;:~`'<>/\|_@#$%^&*()_-+={}[]"    
    if inputString.isalpha(): 
        return 1 # only letters
    elif inputString.isdigit(): 
        return 2 # only numbers
    elif inputString.isalnum():
        return 3 # letters + numbers
    elif all(char in valid_symbols for char in inputString): 
        return 7 # only symbols
    elif (any(char in valid_symbols for char in inputString)) & (any(char.isalpha() for char in inputString)) & (any(char.isdigit() for char in inputString)): 
        return 4 # letters + numbers + symbols
    elif (any(char in valid_symbols for char in inputString)) & (any(char.isalpha() for char in inputString)): 
        return 5 # letters + symbols
    elif (any(char in valid_symbols for char in inputString)) & (any(char.isdigit() for char in inputString)):
        return 6 # numbers + symbols
    else: 
        return 8 
```

```python
def feature_2(inputString):
    num_let = 0
    for char in inputString:
        if char.isalpha(): 
            num_let += 1
    return num_let


def feature_3(inputString):
    num_num = 0
    for char in inputString:
        if char.isdigit(): 
            num_num += 1
    return num_num


def feature_4(inputString):
    num_symb = 0
    valid_symbols = "!?.,;:~`'<>/\|_@#$%^&*()_-+={}[]"    
    for char in inputString:
        if char in valid_symbols: 
            num_symb += 1
    return num_symb
```
