### Задание 1


Имеется статистика за 2016 год по количеству больничных дней для сотрудников разных возрастов и пола. 

Планируется ввести дополнительные медицинские осмотры среди групп риска. Чтобы выявить эти группы необходимо проверить следующие гипотезы: 

1. Мужчины пропускают в течение года более 2 рабочих дней по болезни значимо чаще женщин.
2. Работники старше 35 лет пропускают в течение года более 2 рабочих дней по болезни значимо чаще своих более молодых коллег.



```r
library("broom")
library("ggplot2")
library("mfx")
library("vcd")
library("car")
```

```r
stat16 <- read.csv(file="Статистика за 2016 год.csv", header=TRUE, sep=",", stringsAsFactors = TRUE) # чтобы не было проблем с загрузкой пришлось пересохранить файл с кодировкой "UTF-8"
```

```r
head(stat16)
```

Посмотрим на основные статистики: 

```r
tidy(stat16[1:2])
```

Визуализируем эти данные:

```r
ggplot(stat16) + 
  geom_point(aes(x = Возраст, y = Количество.больничных.дней, color = Пол)) + 
  labs(x = "Возраст, лет", 
       y = "Количество больничных дней",
  title = "Статистика за 2016 год") +
  scale_colour_discrete(name = "Пол")
```

Оценим обыкновенную множественную регрессию для того, чтобы понять, как количество больничных дней зависит от возраста и пола:

```r
lm_model <- lm(Количество.больничных.дней ~ Возраст + Пол + Пол*Возраст, data = stat16) 
summary(lm_model) 
```

```r
ggplot(stat16) + 
  geom_point(aes(x = Возраст, y = Количество.больничных.дней, color = Пол)) +
  geom_abline(slope = lm_model$coefficients[2], intercept = lm_model$coefficients[1], colour='red')  +
  geom_abline(slope = lm_model$coefficients[2] + lm_model$coefficients[4]  , intercept = lm_model$coefficients[1] + lm_model$coefficients[3]  , colour='blue' )  +
  labs(x = "Возраст, лет", 
       y = "Количество больничных дней",
  title = "Статистика за 2016 год") +
  scale_colour_discrete(name = "Пол")
```

Согласно полученным оценкам, мужчины болеют в среднем на 1 день реже, но этот коэффициент  не является значимым.  

Очевидно, множественная регрессия для анализа здесь не подходит, коффициенты в модели незначимы (кроме константы) и достоверных выводов здесь получить мы не сможем.

Перейдем к более подходящей для дискретных объясняемых переменных модели -  логистической регрессии.

Учитывая вопросы, интересующие нас в исследовании, есть смысл перейти к бинарной объяснемой переменной:


$$y_i = \begin{cases} 0, \text{если пропущено не более 2 дней} \\ 1 , \text{если пропущено более 2 дней} \end{cases}$$

```r
stat16['Больничные.дамми']  <- as.numeric(stat16$Количество.больничных.дней > 2) 
```

```r
m_logit <- glm(Больничные.дамми ~ Возраст + Пол ,
          data = stat16,
          family = binomial(link = "logit"))
```

```r
m_logit
```

```r
tidy(m_logit)
```

Все коэффициенты получились  незначимы (на 5% уровне)

Построим полученные регресии для мужчин и для женщин:

```r
temp.data <- data.frame(Возраст=seq(10, 100,len=100), Пол =  sample(c("М","Ж"), replace=TRUE, size=100))
temp.data['Возраст.дамми']  <-  as.factor(as.numeric(temp.data$Возраст > 35))
temp.data$Больничные.дамми = predict(m_logit, newdata=temp.data, type="response")
predicted.data <- as.data.frame(predict(m_logit, newdata = temp.data, type="link", se=TRUE))
new.data <- cbind(temp.data, predicted.data)
```

```r
std <- qnorm(0.95 / 2 + 0.5)
new.data$ymin <- m_logit$family$linkinv(new.data$fit - std * new.data$se)
new.data$ymax <- m_logit$family$linkinv(new.data$fit + std * new.data$se)
new.data$fit <- m_logit$family$linkinv(new.data$fit)  #  rescale to 0-1
```

```r
ggplot(stat16, aes(x=Возраст, y=Больничные.дамми, colour=as.factor(Пол))) + 
    geom_point() + geom_ribbon(data=new.data, aes( ymin=ymin, ymax=ymax, fill=as.factor(Пол)), alpha=0.5) + 
    geom_line(data=new.data, aes(y=fit, colour=as.factor(Пол))) + theme(aspect.ratio = 0.75) +
    labs(x="Temperature", y="Thermal Distress") 
```

Получив две кривые, предсказывающие вероятность, что работник будет болеть более 2 дней (отдельно для мужчин и женщин), можно увидеть, выводы сопостовимые со случаем множественной регресии: с возрастом вероятность заболеть более, чем на 2 дня снижается, вероятность заболеть более, чем на два дня для женщин выше.

Однако стоит отметить, что все полученные коэффициенты оказались незначимыми. Это неудивительно, ведь данные имеющиеся у нас не дают возможности построить кривую регрессию, точно предсказывающую, будет ли работник болеть более двух дней или нет, поскольку болеющие и неболеющие мужчины и женщины распределились примерно поровну (все точки друг над другом, в области около 50% вероятности заболеть). 

Введем также дамми возраста и посмотрим на доверительные интервалы, полученные при оценке логистической регрессии:

$$ x_i = \begin{cases} 0, \text{если не старше 35} \\ 1 , \text{если старше 35} \end{cases}$$

```r
stat16['Возраст.дамми']  <-  as.factor(as.numeric(stat16$Возраст > 35))
```

Чтобы провести анализ модели и увидеть связь между бинарными переменными, можно использовать "мозаичный" график.

```r
mosaic( ~ Больничные.дамми +  Возраст.дамми + Пол  , data = stat16, shade = TRUE)
```

Остатки Пирсона - это индивидуальные вклады в статистику Пирсона  для каждого наблюдения, которые можно использовать для проверки качества модели.  В хорошей модели остатки должны быть небольшими.

Цвет прямоугольников показывает отклонение остатков от ожидаемого значения. Как мы видим на построенной диаграмме, значения остатков Пирсона менее 2, говорит о независимости соответсвующих остатков.

```r
m_logit <- glm(Больничные.дамми ~ Возраст + Возраст.дамми + Пол ,
          data = stat16,
          family = binomial(link = "logit"))
```

```r
m_logit
```

Оценки коэффициентов:

```r
tidy(m_logit)
```

Предельные эффекты:

```r
logitmfx(Больничные.дамми ~ Возраст + Возраст.дамми + Пол ,
          data = stat16)
```

Доверительные интервалы:

```r
require(MASS)
confint(m_logit)
```

Полученные доверительные интервалы указывают на то, что мы не можем быть до конца уверенными даже в знаке влияния переменных на вероятность пропустить по болезни более 2 дней (на 5 % уровне значимости). 

Попробуем получить бутстрэповские доверительные интервалы. 
Вообще с помощью бутспрэпа можно получить  более узкие доверительные интервалы, например, когда данные не соответствуют предположениям модели.
Если результаты бутстрэповских методов существенно отличаются от полученных ранее, то в данных есть проблемы.  

```r
m_logit.boot <- Boot(m_logit, R=1999)
m_logit.boot
```

```r
hist(m_logit.boot, legend="separate")
```

Получим бутстрэповские доверительные интервалы:


```r
Confint(m_logit.boot, level=.95, type="norm")
```

Сравним с результатами, полученными ранее:

```r
confint(m_logit)
```

Полученные бутстрэпом интервалы практически не отличаются, от доверительных интрвалов полученных ранее. Поэтому выводы остаются прежними.

#### Выводы:

Для выявления групп риска, которым требуются дополнительные медицинские осмотры необходимо было проверить следующие гипотезы: 

1. Мужчины пропускают в течение года более 2 рабочих дней по болезни значимо чаще женщин.
2. Работники старше 35 лет пропускают в течение года более 2 рабочих дней по болезни значимо чаще своих более молодых коллег.

Для прверки этих гипотез мы построили логистическую регресиию, получив при этом следующие оценки коэффициентов:


```r
tidy(m_logit)
```

```r
confint(m_logit)
```

Первая гипотеза не подтвердилась, поскольку коэффициент при дамми пола меньше нуля (что значит, что женщины болеют более 2 дней чаще), однако значение p-value > 0.5,  а также достаточно широкий доверительный интервал, говорит о том, что коэффициент статстически незначим и мы даже не можем быть уверены в знаке этого коэффициента.

Вторая гипотеза не подтвердилась, поскольку хотя коэффициент при дамми возраста больше нуля (что значит, что работники старше 35 лет болеют более 2 дней чаще), значение p-value > 0.5,  а также достаточно широкий доверительный интервал, говорит о том, что коэффициент статстически незначим и мы даже не можем быть уверены в знаке этого коэффициента.
